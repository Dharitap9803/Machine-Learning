```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/churn-bigml-80.csv')
```

```python
churn_df.head()
```

Data Manipulation

```python
# sum of missing values
churn_df.isnull().sum()
```

```python
# Categorical values must be converted to numerical labels using label encoder, one-hot encoding,
# Label Encoder is used for mapping categorical values to 0, 1, 2, 3, 4 ...
from sklearn.preprocessing import LabelEncoder
le_encoder = LabelEncoder()
```

```python
le_encoder.fit(churn_df['International plan'])
```

```python
churn_df['International plan'] = le_encoder.transform(churn_df['International plan'])
```

```python
le_encoder.fit(churn_df['Voice mail plan'])
```

```python
churn_df['Voice mail plan'] = le_encoder.transform(churn_df['Voice mail plan'])
```

Correlation

```python
plt.figure(figsize = (20,15))
sns.heatmap(churn_df.corr(), annot = True)
```

Imbalanced Class Data

```python
# Imbalanaced Class Data
sns.catplot(x = 'Churn', kind = 'count', data = churn_df, palette = 'magma', height = 4)
```

```python
churn_df['Churn'].value_counts()
```

Data Augmentation

```python
from sklearn.utils import resample
df_0 = churn_df[churn_df['Churn'] == False]
df_1 = churn_df[churn_df['Churn'] == True]
```

```python
# Apply Resample
df_1_upsample = resample(df_1, n_samples = 1139, replace = True, random_state = 123)
```

```python
# Merge
churn_df1 = pd.concat([df_0, df_1_upsample])
```

```python
churn_df1['Churn'].value_counts() / len(churn_df1['Churn'])
```

Features & Target

```python
churn_df1.columns
```

```python
X = churn_df1.drop(['State','Number vmail messages','Total day charge','Total eve charge','Total night charge','Total intl charge',
                    'Churn'], axis = 1)
Y = churn_df1['Churn']
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
# class_weight is useful in class weight distribution
clf_tree = DecisionTreeClassifier(criterion = 'entropy', class_weight = {0: 66.67, 1: 33.33}, max_depth = 11)
```

```python
clf_tree.fit(x_train, y_train)
```

```python
clf_tree.score(x_test, y_test)
```

```python
clf_tree.score(x_train, y_train)
```

```python
clf_tree.get_depth()
```

Plot Decision Tree

```python
from sklearn import tree
fig = plt.figure(figsize = (130, 30))
_ = tree.plot_tree(clf_tree, feature_names = X.columns, class_names = str(churn_df1.Churn), filled = True,
               rounded = True, fontsize = 9)
```

Classification tree

```python
predictions = clf_tree.predict(x_test)
```

Classification Metrices

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, predictions)
```

```python
plt.figure(figsize = (4,3))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='magma', fmt = '0.0f')
```

```python
print(classification_report(y_test, predictions))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# bootstrap = random sampling of existing features, n_estimators = Number of Decision Trees
clf_rf = RandomForestClassifier(bootstrap = True, n_estimators = 100, criterion = 'entropy', max_depth = 10)
```

```python
clf_rf.fit(x_train, y_train)
```

```python
clf_rf.score(x_test, y_test)
```

```python
clf_rf.score(x_train, y_train)
```

```python
predictions = clf_rf.predict(x_test)
```

```python
plt.figure(figsize = (4,3))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='magma', fmt = '0.0f')
```

```python
print(classification_report(y_test, predictions))
```
